```python
# Dependencies and Setup
import pandas as pd

# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_schools = school_data_complete.groupby(['school_name']).ngroups
#total_schools = len(school_data['school_name'].unique())
#total_schools = school_data['school_name'].nunique()

#Calculate the total number of students
total_students= school_data_complete.groupby(['Student ID']).ngroups
#total_students = len(student_data['Student ID'].unique())
#total_schools = student_data['Student ID'].nunique()

#Calculate the total budget
total_budget= school_data['budget'].sum()

#Calculate the average math score
avg_math_score = school_data_complete['math_score'].mean()

#Calculate the average reading score
avg_reading_score = school_data_complete['reading_score'].mean()

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
passing_rate = (avg_math_score+avg_reading_score)/2

#Calculate the percentage of students with a passing math score (70 or greater)
good_math = student_data.loc[student_data['math_score']>=70]
count_good_math = good_math['Student ID'].nunique()
percent_math_over70 = (count_good_math*100)/total_students

#Calculate the percentage of students with a passing reading score (70 or greater)
good_reading = student_data.loc[student_data['reading_score']>=70]
count_good_reading = good_reading['Student ID'].nunique()
percent_reading_over70 = (count_good_reading*100)/total_students

#Create a dataframe to hold the above results. 
summary_df = pd.DataFrame({
    "Total Schools":[total_schools],
     "Total Students":[total_students],
      "Total Budget":[total_budget],
      "Average Math Score":[avg_math_score],
      "Average Reading Score":[avg_reading_score],
      "% Passing Math":[percent_math_over70],
      "% Passing Reading":[percent_reading_over70],
      "% Overall Passing Rate":[passing_rate]})

#Optional: give the displayed data cleaner formatting
summary_df["Total Students"] = summary_df["Total Students"].map("{:,}".format)
summary_df["Total Budget"] = summary_df["Total Budget"].map("${:,}".format)
summary_df["Average Math Score"] = summary_df["Average Math Score"].map("{:.3f}".format)
summary_df["Average Reading Score"] = summary_df["Average Reading Score"].map("{:.3f}".format)
summary_df["% Passing Math"] = summary_df["% Passing Math"].map("{:.2f}%".format)
summary_df["% Passing Reading"] = summary_df["% Passing Reading"].map("{:.2f}%".format)
summary_df["% Overall Passing Rate"] = summary_df["% Overall Passing Rate"].map("{:.2f}%".format)
summary_df.head()
```

# School Summary

```python
# School Summary: Create an overview table that summarizes key metrics about each school, including:

#School Type
school_type = school_data.set_index(['school_name'])['type']
#school_type

#Total Students per school
total_school_students = school_data_complete.groupby('school_name')['Student ID'].nunique()
#total_school_students

#Total School Budget
total_school_budget = school_data.groupby('school_name')['budget'].sum()
#total_school_budget

#Per Student Budget
per_student_budget = total_school_budget/total_school_students
#per_student_budget

#Average Math Score
avg_math_score_school = school_data_complete.groupby('school_name')['math_score'].mean()
#avg_math_score_school

#Average Reading Score
avg_reading_score_school = school_data_complete.groupby('school_name')['reading_score'].mean()
#avg_reading_score_school

#% Passing Math
good_math_school = school_data_complete.loc[student_data['math_score']>=70]
count_good_math_school = good_math_school.groupby('school_name')['Student ID'].nunique()
percent_math_over70_school = (count_good_math_school*100)/total_school_students

#% Passing Reading
good_reading_school = school_data_complete.loc[student_data['reading_score']>=70]
count_good_reading_school = good_reading_school.groupby('school_name')['Student ID'].nunique()
percent_reading_over70_school = (count_good_reading_school*100)/total_school_students

#Overall Passing Rate (Average of the above two)
passing_rate_school = (percent_math_over70_school+percent_reading_over70_school)/2
#passing_rate_school


#Create a dataframe to hold the above results
school_summary_df1 = pd.DataFrame({
   "School Type":school_type,
   "Total Students":total_school_students,
   "Budget per Student": per_student_budget,
   "Total School Budget":total_school_budget,
   "Average Math Score":avg_math_score_school,
   "Average Reading Score":avg_reading_score_school,
   "% Passing Math":percent_math_over70_school,
   "% Passing Reading":percent_reading_over70_school,
   "% Overall Passing Rate":passing_rate_school
    })

school_summary_df = pd.DataFrame({
   "School Type":school_type,
   "Total Students":total_school_students,
   "Budget per Student": per_student_budget,
   "Total School Budget":total_school_budget,
   "Average Math Score":avg_math_score_school,
   "Average Reading Score":avg_reading_score_school,
   "% Passing Math":percent_math_over70_school,
   "% Passing Reading":percent_reading_over70_school,
   "% Overall Passing Rate":passing_rate_school
    })
school_summary_df

school_summary_df["Total Students"] = school_summary_df["Total Students"].map("{:,}".format)
school_summary_df["Budget per Student"] = school_summary_df["Budget per Student"].map("${:,}".format)
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,}".format)
school_summary_df["Average Math Score"] = school_summary_df["Average Math Score"].map("{:.3f}".format)
school_summary_df["Average Reading Score"] = school_summary_df["Average Reading Score"].map("{:.3f}".format)
school_summary_df["% Passing Math"] = school_summary_df["% Passing Math"].map("{:.2f}%".format)
school_summary_df["% Passing Reading"] = school_summary_df["% Passing Reading"].map("{:.2f}%".format)
school_summary_df["% Overall Passing Rate"] = school_summary_df["% Overall Passing Rate"].map("{:.4f}%".format)
school_summary_df.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Grouping the dataframe by the grade
school_math_9th = school_data_complete[school_data_complete['grade']=='9th']
school_math_9th_df = school_math_9th.groupby('school_name')['math_score'].mean()

school_math_10th = school_data_complete[school_data_complete['grade']=='10th']
school_math_10th_df = school_math_10th.groupby('school_name')['math_score'].mean()

school_math_11th = school_data_complete[school_data_complete['grade']=='11th']
school_math_11th_df = school_math_11th.groupby('school_name')['math_score'].mean()

school_math_12th = school_data_complete[school_data_complete['grade']=='12th']
school_math_12th_df = school_math_12th.groupby('school_name')['math_score'].mean()

#math score according to each grade
math_scores_grade = pd.DataFrame({
    '9th':school_math_9th_df.map("{:.3f}".format),
    '10th':school_math_10th_df.map("{:.3f}".format),
    '11th':school_math_11th_df.map("{:.3f}".format),
    '12th':school_math_12th_df.map("{:.3f}".format)
})
math_scores_grade
```

## Reading Score by Grade

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins
bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# Combining a dataframe for required columns for the solution
budget_per_student = school_summary_df1.loc[:,['Average Math Score',
                                  'Average Reading Score','% Passing Math',
                                  '% Passing Reading','% Overall Passing Rate',]]
# create a new column named Spending Ranges (Per Student) and slicing the data
budget_per_student['Spending Ranges (Per Student)']= pd.cut(school_summary_df1['Budget per Student'],bins,labels=group_names)
# Create a group based on the bins calculating the mean
budget_per_student = budget_per_student.groupby('Spending Ranges (Per Student)').mean()
budget_per_student.head()
```
